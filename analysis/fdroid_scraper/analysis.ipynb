```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from urllib.parse import urlparse
```

```python
df_fdroid = pd.read_csv("output/fdroid_apps_list_complete.csv")
```

```python
df_fdroid.head()
```

```python
def extract_domain(url):
    if isinstance(url, str):
        return urlparse(url).netloc
    return "NO DOMAIN"
```

```python
STATS = ["source_code", "issue_tracker"]

for stat in STATS:
    # http domains
    domains = df_fdroid[stat].apply(extract_domain)
    domains_values = domains.value_counts().rename_axis('domain').reset_index(name='counts')
    domains_values["percentage"] = domains_values["counts"]
    domains_values["percentage"] = domains_values["percentage"].apply(lambda x: x / domains.shape[0])
    domains_values.to_csv(f"output/{stat}_counts.csv", index=None)
```
